```python
# Importing required modules
import torch
from transformers import BertTokenizer, BertModel
```

```python
# Importing required modules
import logging
import matplotlib.pyplot as plt
```

```python
# Loading model tokenizer(pre-trained)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Random text for testing
text = "BERT is being used to generate embeddings of the input text."
# Adding the [CLS] and [SEP] markers
marked_text = "[CLS] " + text + " [SEP]"
```

```python
# Tokenizing the random text for testing purpose
tokenized_text = tokenizer.tokenize(marked_text)
print (tokenized_text)
```

```python
# Random text used for testing 
text = "As he crossed toward the pharmacy at the corner he involuntarily turned "\
       "his head because of a burst of light that had ricocheted from his temple, "\
       "and saw, with that quick smile with which we greet a rainbow or a rose, a "\
       "blindingly white parallelogram of sky being unloaded from the van."
    
marked_text = "[CLS] " + text + " [SEP]"
```

```python
# Tokenizing the text for testing BERT
tokenized_text = tokenizer.tokenize(marked_text)
```

```python
# Map the token strings to their vocabulary indeces
indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
```

```python
# Displaying the words in the text with their indices
for tup in zip(tokenized_text, indexed_tokens):
    print('{:<12} {:>6,}'.format(tup[0], tup[1]))
```

```python
# Marking each of the tokens as belonging to sentence "1"
segments_ids = [1] * len(tokenized_text)
print (segments_ids)
```

```python
# Converting input text to Pytorch tensors
tokens_tensor = torch.tensor([indexed_tokens])
segments_tensors = torch.tensor([segments_ids])
```

```python
# Load pre-trained model
model = BertModel.from_pretrained('bert-base-uncased',
                                  output_hidden_states = True,
                                 )
# Putting the model in "evaluation" mode
model.eval()
```

```python
# Feeding the random text to the BERT model for generating hidden states (embeddings)
with torch.no_grad():
  outputs = model(tokens_tensor, segments_tensors)
  hidden_states = outputs[2]
```
